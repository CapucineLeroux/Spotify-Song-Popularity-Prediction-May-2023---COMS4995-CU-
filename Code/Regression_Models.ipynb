# PART A - Initial Models

# 1. Preliminary imports

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import requests
import base64
import json
import seaborn as sns

from category_encoders import TargetEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import zipfile
```

# 2. Data preprocessing

```python
# Load the cleaned dataset
with zipfile.ZipFile('../Data/clean_df.zip', 'r') as zip_ref:
    zip_ref.extractall("../Data")
df = pd.read_csv('../Data/clean_df.csv')
print("Clean dataset has "+str(df.shape[0])+" samples, "+str(df.shape[1])+" features.")
```

```python
# Remove duplicate songs
duplicate_rows = df.duplicated(keep='first', subset=["name_tracks", "name_artists"])
duplicate_rows = ~duplicate_rows
df = df[duplicate_rows]
df.shape
```

#### **2.1. Encode the categorical features (artist name, artist id, genres)**

2.1.1. Do an ordinal encoding on the name ('name_artists') and the id ('id_artists') of the artists.

```python
# Map id and name of artists to a single numerical id
unique_artists = df[['name_artists', 'id_artists']].drop_duplicates()
artist_ids, unique_values = pd.factorize(unique_artists['id_artists'])
unique_artists['artist_num_id'] = artist_ids
df = pd.merge(df, unique_artists[['name_artists', 'artist_num_id']], on='name_artists', how='left')

# Drop unused columns 
dropped_columns = ['Unnamed: 0','id_tracks','name_tracks','id_artists','name_artists']
df = df.drop(columns=dropped_columns)
```

2.1.2. Do a target encoding on genres. 

```python
# Transform the string of genres to a tuple
def genre_transform (genre_string):
  genre_list = genre_string.split(", ")
  genre_list = [s.strip("'") for s in genre_list]
  return tuple(genre_list)

df['genres'] = df['genres'].apply(genre_transform)

# Print the number of genres
all_genres = np.unique(np.concatenate(df["genres"].values))
print("There are "+str(len(all_genres))+" different genres in the dataset.")

# Do a target encoding on each tuple of genres
encoder = TargetEncoder(cols=["genres"])
df_target = df['popularity_tracks']
encoder.fit(df, df_target)
df = encoder.transform(df)
```

*There are 4407 different genres, it would have been impossible to do a one-hot encoding, this is why we used the target encoding instead. Each set of genres is replaced by its average track popularity.*

#### **2.2. Remove highly collinear features by reading the correlation plot.**

```python
df_X = df.drop(columns=["popularity_tracks"])
df_y = df["popularity_tracks"]

corr = df_X.corr()
sns.heatmap(corr)

#Find the pairs of features correlated above 0.9
corr_column_names = corr.columns
highly_correlated_pairs = []

for i in range (0,len(corr_column_names)):
    corr_column_name = corr_column_names[i]
    corr_column_values = corr[corr_column_name].values
    for j in range (i,len(corr_column_values)):
        if (i!=j) and (abs(corr_column_values[j]) >= 0.9):
            highly_correlated_pairs.append((corr_column_names[i],corr_column_names[j]))

print("Highly correlated pairs = "+str(highly_correlated_pairs))
#Remove the first column of each pair
for pair in highly_correlated_pairs:
    if pair[0] in df_X.columns:
        df_X = df_X.drop(pair[0],axis=1)
        print("Removed column "+pair[0])
```

#### **2.3. Split the dataset into development (80%) and test (20%) using a random split.**

```python
X = df.drop(columns=["popularity_tracks"])
y = df["popularity_tracks"]
X_dev, X_test, y_dev, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print("Sizes:\nX_dev: "+str(len(X_dev))+"\nX_test: "+str(len(X_test)))
```

#### **2.4. Standardize the columns in the feature matrices X_dev and X_test to have zero mean and unit variance.**

```python
scaler = StandardScaler()
X_dev = scaler.fit_transform(X_dev)   # Fit and transform scalar on X_dev
X_test = scaler.transform(X_test)         # Transform X_test
```

# 3. Training

#### **3.1. Linear Regression**

```python
# Training performance
linear_model = LinearRegression()
scores = cross_val_score(linear_model, X_dev, y_dev, cv=10, error_score="raise")
print('Cross validation scores:', scores)
print('Mean cross validation error:', scores.mean())

# Test performance
linear_model.fit(X_dev, y_dev)
test_score = linear_model.score(X_test, y_test)
print('\nTest score:', test_score)
```

```python
# Plot the different coefficients to interpret feature importance
fig = plt.figure(figsize=(8,5))
ax = sns.barplot(x=X.columns, y=linear_model.coef_)
ax.tick_params(axis='x',rotation=90)
plt.xlabel("Feature")
plt.ylabel("Model's Coefficient")
plt.title("Feature importance")
plt.show()
```

```python
# Hyperparameter tuning the ridge model
param_grid = {'alpha': np.logspace(-3, 3, 10)}
grid_search = GridSearchCV(ridge_model, param_grid, return_train_score=True)
grid_search.fit(X_dev, y_dev)

# Get the best hyperparameters and print them
best_params = grid_search.best_params_
best_score = grid_search.best_score_
print("Best hyperparameters:", best_params)
print("Best score:", best_score)

# Create a Ridge model object with the best hyperparameters
tuned_ridge_model = Ridge(alpha=best_params['alpha'])
tuned_ridge_model.fit(X_dev, y_dev)

# Test peformance
test_score = tuned_ridge_model.score(X_test, y_test)
print('\nTest score:', test_score)
```
